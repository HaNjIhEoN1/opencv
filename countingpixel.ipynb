```python
def findgreen(img):
    raw = mpimg.imread(img, cv2.IMREAD_COLOR)
    lower_green = (0, 51, 0)
    upper_green = (90, 204, 90)
    img_mask = cv2.inRange(raw, lower_green, upper_green)
    img_result = cv2.bitwise_and(raw, raw, mask=img_mask)
    gray = cv2.cvtColor(img_result, cv2.COLOR_BGR2GRAY)
    border, binary = cv2.threshold(gray,60 , 255, cv2.THRESH_BINARY)
    nzCount = cv2.countNonZero(binary)
    return nzCount
```

```python
findgreen('test.jpg')
```
